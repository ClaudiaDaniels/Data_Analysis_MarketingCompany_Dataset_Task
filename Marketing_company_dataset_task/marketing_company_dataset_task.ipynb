## CHECK FOR MISSING VALUES

```python
datafill_zero= chart_assignment.fillna(0)
datafill_zero
```

## Handle missing values

```python
datafill=chart_assignment.fillna(chart_assignment['Impressions']. mean())
datafill=chart_assignment.fillna(chart_assignment['Clicks']. mean())
chart_asssignment=datafill
print(chart_assignment)
```

## Format Data Types 

```python
chart_assignment['Date']= pad.to_datetime(chart_assignment['Date'])
chart_assignment
```

## Bar Chart: the total number of conversions by region. 

```python
sns.barplot(x='Region', y= 'Conversions', data= chart_assignment)
plat.title ('chart_assignment Bar chart')
plat.show
```

## Changing the color of the bars."

```python
sns.barplot(x='Region', y= 'Conversions', data= chart_assignment, color= 'red')
plat.title ('chart_assignment Bar chart color change')
plat.show
```

## Scatter Plot: scatter plot of Clicks vs Impressions 

```python
Impressions = [13116.0, 25470.0, 403440.0, 50573.0, 56573.0, 65120.0, 95747.0]  # Impressions
Clicks = [490.0, 1593.0, 2059.0, 2585.0, 3100.0, 4510.0, 4948.0]  # Clicks
plat.scatter(Impressions, Clicks, color='green', label='Data Points')
z = nup.polyfit(Impressions, Clicks, 1)  # 1 indicates a linear fit (degree 1)
p = nup.poly1d(z)
plat.plot(Impressions, p(Impressions), color='red', linestyle='--', label='Trendline')
plat.xlabel('Impressions')
plat.ylabel('Clicks')
plat.title('Scatter Plot of Clicks vs Impressions with Trendline')
plat.legend()
plat.show()
```

## Pie Chart: the percentage of total budget allocation across the four regions

```python
Budget_count= chart_assignment['Region']. value_counts()
plat.pie(Budget_count, labels = Budget_count.index,autopct='%1.1f%%', colors=sns.color_palette('Set2'))
plat.title('Region % Distribution')
plat.show()
```

##  Line Chart: the total number of impressions over time (using the Date column)

```python
sns.lineplot(data=chart_assignment, x= 'Impressions', y= 'Date', color= 'blue')
plat.xlabel('x axis label')
plat.ylabel('y axis label')
plat.title ('chart_assignment line chart')
plat.show()
```

## Histogram: the distribution of Budget (USD) values."

```python
sns.histplot(chart_assignment['Budget (USD)'],bins=10,kde=True, color = 'purple')
plat.title ('Budget (USD) Distribution')
plat. xlabel ('Budget')
plat. ylabel
plat. show()
```

## Trendline: a scatter plot of Budget (USD) vs Conversions, and a trendline analyzing the relationship."


```python

Budget = [1000,2000,3000,4000,5000,6000,7000,8000,9000,10000] #budget in USD
Conversions = [50,100,150,200,250,300,350,400,450,500] #Number of conversions
plat.scatter(Budget,Conversions, color='blue', label='data points')
z = nup.polyfit (Budget, Conversions, 1) #1 indicates a linear fit (degree 1)
p = nup.poly1d (z)
plat.plot(Budget, p(Budget), color='blue', linestyle='--', label='Trendline')
plat.xlabel('Budget (USD)')
plat.ylabel('Conversions')
plat.title('Scatter Plot of Budget vs Conversions with Trendline')
plat.legend()
plat.show()
```

## Clustered Bar Chart: a clustered bar chart comparing the number of clicks and conversions by region."

```python
plat. figure(figsize=(10, 6))
sns.barplot(x='Conversions', y= 'Clicks', hue='Region', data= chart_assignment, palette= "muted")
plat.title ('chart_assignment Bar chart')
plat. xlabel('Conversions')
plat. ylabel('Clicks')
plat.show()
```

## Boxplot: a boxplot visualizing the spread of Impressions values by region. each box are coloured differently."

```python
sns.boxplot(x='Impressions', y='Region', data=chart_assignment, palette= 'tab10')
plat.show()
```

## Heatmap: a correlation heatmap showing the relationship between Budget, Clicks, Impressions, and Conversions."

```python
sns.heatmap(chart_assignment[['Budget (USD)', 'Clicks', 'Impressions', 'Conversions']].corr(),annot=True, cmap='coolwarm')
plat.title('chart_assignment heatmap')
plat.show()
```

## 3D Plot: A 3D scatter plot using Clicks, Impressions, and Conversions. points coloured based on the region."
